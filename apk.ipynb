```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data-mall.csv')
```

```python
df.head()
```

```python
df.rename(index=str, columns={
    'Annual Income (k$)' : 'Income',
    'Spending Score (1-100)' : 'Score'
}, inplace = True)
```

```python
df.head()
```

```python
X = df.drop(['CustomerID', 'Gender'], axis=1)
```

```python
print(X)
```

```python
, ConnectionStyle
from matplotlib.patches import ArrowStyle
clusters=[]
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)

fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(x=list(range(1,11)),y=clusters, ax=ax)
ax.set_title('pencarian elbow')
ax.set_xlabel('cluster')
ax.set_ylabel('inertia')

#panah elbow
ax.annotate('Possible elbow point', xy=(3, 1900000), xytext=(3, 50000), xycoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))

ax.annotate('Possible elbow point', xy=(5, 80000), xytext=(2, 2500000), xycoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))
```

```python
#klaster dibagi menjadi 4 klaster
n_clust = 3
kmean = KMeans(n_clusters=n_clust).fit(X)
X['Labels'] = kmean.labels_
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X['Income'], y=X['Score'], hue=X['Labels'], markers=True, size=X['Labels'], palette=sns.color_palette('hls', n_clust))

#keterangan
for label in X['Labels']:
  plt.annotate(label,
               (X[X['Labels']==label]['Income'].mean(),
                X[X['Labels']==label]['Score'].mean()),
                horizontalalignment ='center',
                verticalalignment='center',
                size=20, weight='bold',
                color='black')
```
